### 주가 데이터프레임

```python
juga=pd.read_csv('주가 크롤링.csv')
```

```python
juga.head()
```

```python
juga.isna().sum()
```

```python
lists=[]
for i in range(len(juga['등락구분'])):
    if juga.iloc[i]['등락구분']==True:
        lists.append(1)
    else: lists.append(-1)
        
juga['label']=lists
```

```python
juga
```

### 뉴스데이터보다 하루 뒤에 주가데이터가 반영되므로 주가데이터 (날짜-1)일을 하여 데이터 프레임을 합쳐준다

```python
from datetime import timedelta
```

```python
list1=[]
list2=[]
for i in range(len(juga['Date'])):
    time=juga.iloc[i]['Date']
    times=time.replace('00:00:00','')
    times=times.replace(' ','')
    times=datetime.datetime.strptime(times,'%Y-%m-%d')
    list1.append(times)
    times=times-timedelta(days=1)
    list2.append(times)
juga['주가 반영 Date']=list1
juga['Date']=list2 #dataframe 합치기위한 가상의 date=실제date-1
```

```python
df=pd.merge(data,juga,on='Date',how='outer')
```

# 데이터프레임 확인
